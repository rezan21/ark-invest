```python
import pandas as pd
from pathlib import Path
```

```python
files = list(Path("./ETFs/").glob("*.csv"))
df = [pd.read_csv(f) for f in files]
df = pd.concat(df, ignore_index=True)
df.rename({"market value($)":"value", "weight(%)":"weight"}, axis=1, inplace=True)
df = df.dropna(subset=["fund", "company", "ticker", "cusip", "shares"], how="all").reset_index(drop=True)
df
```

```python
df.groupby(["ticker", "company"], as_index=False).agg({"value":"sum"}).sort_values("value", ascending=False)
```

```python
df.groupby(["ticker", "company"], as_index=False).agg({"weight":"sum"}).sort_values("weight", ascending=False)
```
