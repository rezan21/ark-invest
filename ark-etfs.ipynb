```python
from pathlib import Path

import pandas as pd
```

```python
ETF = "ARKG"
etf_his_files = list(Path(f"./ETFs/{ETF}/").glob("*.csv"))
df = pd.concat([pd.read_csv(f).dropna() for f in etf_his_files])
df
```

```python
df["date"] = pd.to_datetime(df["date"], format="%m/%d/%Y")
df.sort_values(["ticker", "date"], inplace=True)
df
```

```python
gg = df.groupby("ticker")
for g in gg.groups:
    gdf = gg.get_group(g)
```
